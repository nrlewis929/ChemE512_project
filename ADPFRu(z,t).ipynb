```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
# Constants (Table 5)
T = 650 +273.15 # K, 650 C
P = 101325 # Pa, 1 atm
y_in = 0.15 # also use 0.45, 0.75, 0.9
F_in = 20 # NmL/min
F_in = F_in/(1e6*60) # mL/s
d_reactor = 7e-3 # m
d_particle = 112.5e-6 # m
# gam = 0.24
rho0_particle = 1693 #kg/m3
eps0_particle = 0.4
ks = 5.95e-7 # m4/kmol-s
# ks = ks/1e3 # m4/mol-s
alpha = 1.65
a = 22.7
b = 0.35
d0_CaO = 69.5e-9 # m 
Dr = 1e-5 # m2/s
eps = 0.5
rho_CaO = 3340 #kg/m3

# Additional constants (from paper or known)
R = 8.314 #J/molK
R = R * 1000 # J/kmolK
zeta = 2.18 # CaCO3:CaO molar volume ratio
MW_CaO = 56.0774 # g/mol, kg/kmol
H = 0.015 # m 
```

```python
# Computed constants
S = np.pi*(d_reactor/2)**2 #m2, reactor cross sectional area
Ctot = P/(R*T)
u_in = F_in / S
tau_b = H / u_in
v_CaO = MW_CaO/rho_CaO #m3/kmol
n_CaO = rho0_particle / MW_CaO # kmol/m3

# Dimensionless quantities
DR = Dr * tau_b / H**2
U_in = 1 # Dedimensionalized U

# Coefficients from Table 4
C1 = tau_b / n_CaO
C2 = (zeta - 1) * v_CaO * n_CaO
C3 = (1 - eps) * tau_b / Ctot 
C4 = (1 - eps) * C1 * C2 - C3
C5 = (zeta - 1) * v_CaO * Ctot

# additional computed constants
sigma_CaO = 3*(d0_CaO)**2/(d_particle/2)**3 #1/m, surface to volume ratio, double check
print('sigma:',sigma_CaO)
y_eq = 0.0096 # y of CO2 at equilibrium
D_PL0 = 1e-8 # product layer diffusivity coeff, same OEM as Dr
```

```python
# Model functions
def A(X):
    return eps + (1 - eps)*(eps0_particle - C2 * X)

def rcbn(y,X):
    numerator = sigma_CaO * n_CaO * ks * (1 - X)**(2/3) * Ctot * (y - y_eq)
    
    denominator = 1 + (n_CaO * ks)/(2*D_PL0*np.exp(-a*X**b)) * d0_CaO * (1-X)**(1/3) * (1 - ((1-X)/(1-X+X*zeta))**(1/3))
    return numerator/denominator
```

```python
# Define model

def model(y,t):
    dydt = 0*y
#     y[N]=y[N-1]
    y[0] = (y_in+DR/dz*y[1])/(1+DR/dz)
    
    u = 0*y
    u[0] = U_in
    for i in range(1,N+1):
        u[i] = u[i-1] + dz*C4*rcbn(y[i],y[N+i+1])
    
    # initial term
    term1a = (2*DR/(A(y[N+1])*dz**2)) * (y[1] - y[0]) 
    term2a = -(2/(A(y[N+1])*dz) + u[0]/(A(y[N+1])*DR)) * (u[0]*y[0] - U_in*y_in)
    term3a = -(1-y[0]) * C3/A(y[N+1]) * rcbn(y[0],y[N+1])
    dydt[0] = term1a + term2a + term3a
    
    # middle terms
    for j in range(1, N):
        term1b = (DR/(A(y[N+j+1])*dz**2) - u[j]/(2*A(y[N+j+1])*dz))*y[j+1]
        term2b = -(2*DR/(A(y[N+j+1])*dz**2))*y[j]
        term3b = (DR/(A(y[N+j+1]) * dz**2) + u[j]/(A(y[N+j+1])*2*dz)) * y[j-1]
        term4b = -(1-y[j]) * C3/A(y[N+j+1]) * rcbn(y[j],y[N+j+1])
    
        dydt[j] = term1b + term2b + term3b + term4b
    
    # final term
    term1c = (2*DR/(A(y[N+N+1])*dz**2))*(y[N-1] - y[N])
    term2c = -(1-y[N])*C3/A(y[N+N+1])*rcbn(y[N],y[N+N+1])
    dydt[N] = term1c + term2c
    
    # dXdt
    for k in range(N+1,N+N+2):
        dydt[k] = C1 * rcbn(y[k-N-1],y[k])
    
    return dydt
```

```python
# Discretize
N = 100
z = np.linspace(0,1,N+1)
t = np.linspace(0,2,1000)

dz = np.diff(z)[0]
dt = np.diff(t)[0]
```

```python
# results using combined model
y0 = np.zeros(N+N+2)
# X = 0 leads to dividing by 0, so make it really small
y0[-N:] = y0[-N:]+ 0.0000000001

result = odeint(model, y0, t)

Y = result[:,:101]
X = result[:,101:]
```

```python
# Back out u[i]
u = np.zeros((len(t),N+1))
u[:,0] = U_in
for i in range(1,N+1):
    u[:,i] = u[:,i-1] + dz*C4*rcbn(Y[:,i-1],X[:,i-1])
   
```

```python
Z, T = np.meshgrid(z, t)

# fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
fig = plt.figure(figsize=(12,6))#)plt.figaspect(0.5))

# y plot
ax = fig.add_subplot(1, 3, 1, projection='3d')
surf = ax.plot_surface(Z[:,1:],T[:,1:],Y[:,1:], cmap=cm.coolwarm,
                       linewidth=1, antialiased=False)

# ax.zaxis.set_major_formatter('{x:.02f}')
ax.set_xlabel(r'$z^*$',size=12)
ax.set_ylabel(r'$t [s]$',size=12)
ax.set_title(r'$y_{CO_2}$',size=14)

# X plot
ax = fig.add_subplot(1, 3, 2, projection='3d')
surf = ax.plot_surface(Z[:,1:],T[:,1:],X[:,1:], cmap=cm.coolwarm,
                       linewidth=1, antialiased=False)
ax.set_xlabel(r'$z^*$',size=12)
ax.set_ylabel(r'$t [s]$',size=12)
ax.set_title(r'$X$',size=14)

# U plot
ax = fig.add_subplot(1, 3, 3, projection='3d')
surf = ax.plot_surface(Z,T,u, cmap=cm.coolwarm,
                       linewidth=1, antialiased=False)
ax.set_xlabel(r'$z^*$',size=12)
ax.set_ylabel(r'$t [s]$',size=12)
ax.set_title(r'$u$',size=14)

# plt.tight_layout()
plt.suptitle('ADPFRu(z,t)',y=.85,size=16)
plt.show()
```
